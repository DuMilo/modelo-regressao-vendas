```python
import pandas as pd
import sklearn as sk
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
# Sklearn Libraries

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# Data

dados_vendas = {
    'mes': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],
    'vendas': [2000, 2200, 2300, 2500, 2600, 2700, 2800, 2900, 3000, 3100, 3200, 3300]
}
```

### Tratamento de Dados

```python
df_vendas = pd.DataFrame.from_dict(dados_vendas)
```

```python
df_vendas
```

```python
df_vendas['mes_numero'] = range(1, 13)
```

```python
df_vendas
```

```python
X = pd.DataFrame(df_vendas['mes_numero'])
y = df_vendas['vendas']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
model = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.coef_ # a
```

```python
model.intercept_# b
```

```python
# R2 Score no Conjunto de Treinamento

model.score(X_train, y_train) # Modelo Bom
```

```python
# R2 Score no Conjunto de Teste

model.score(X_test, y_test) # Modelo Bom
```

```python
# MSE

mean_absolute_error(y_test, y_pred)
```
